```python
#Time Resampling
```

```python
df_tesla.head()
```

```python
df_tesla.set_index('Date', inplace=True)
```

```python
df_tesla.resample(rule='A').min()
```

```python
#yearly max
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
#quarterly max
df_tesla.resample(rule='QS').max()
```

```python
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
##Business end frequency
df_tesla.resample(rule='QS').max()
```

```python
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').mean().plot(kind='bar',figsize=(15,6))
```

```python
df_tesla.head()
```

```python
#rolling inbuilt function
df_tesla['High'].rolling(10).mean().head(20)
```

```python
df_tesla["Open:30 days roling"]= df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla[['Open','Open:30 days roling']].plot(figsize=(12,5))
```

#Finding jump and market volatility

```python
df_msoft['Volatile'] = df_msoft['High']-df_msoft['Low']
```

```python
df_msoft['Volatile'].idxmax()
```

```python
df_msoft['Volatile : 30 day']= df_msoft['Volatile'].rolling(30).mean()
```

```python
df_msoft[['Volatile', 'Volatile : 30 day']].plot(figsize=(15,6))
```
